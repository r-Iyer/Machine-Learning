```python
number=LabelEncoder()
data['Age']=number.fit_transform(data['Age'].astype('float64'))
number=LabelEncoder()
data['Fare']=number.fit_transform(data['Fare'].astype('float64'))
number=LabelEncoder()
data['Sex']=number.fit_transform(data['Sex'].astype('float64'))
number=LabelEncoder()
data['sibsp']=number.fit_transform(data['sibsp'].astype('float64'))
number=LabelEncoder()
data['Parch']=number.fit_transform(data['Parch'].astype('float64'))
number=LabelEncoder()
data['Pclass']=number.fit_transform(data['Pclass'].astype('float64'))
number=LabelEncoder()
data['Embarked']=number.fit_transform(data['Embarked'].astype('float64'))
number=LabelEncoder()
data['survived']=number.fit_transform(data['survived'].astype('float64'))
```

```python
X=pd.DataFrame(data[['Age','Fare','Sex','sibsp','Parch','Pclass','Embarked']])
Y=data['survived']
```

```python
plt.figure(figsize=(12,10))
cor = pd.concat([X,Y],axis=1).corr()
#sns.heatmap(cor[np.abs(cor)>.8], annot=True, cmap=plt.cm.Reds)
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=8)
```

```python
X_train.shape
```

```python
acc=[]
import warnings
warnings.filterwarnings("ignore")
for i in range(-5,16,2):
    svm = SVC(kernel='linear',C=pow(2,i))
#     svm = SVC(kernel='poly',C=pow(2,i),gamma='auto_deprecated',probability=True)  
    svm.fit(X_train, Y_train)
    Y_pred_svm = svm.predict(X_test)
    acc_svm= round(svm.score(X_test, Y_test) * 100, 2)
    print("SVM",acc_svm," c= ",i)
    acc.append(acc_svm)
    
   # print(confusion_matrix(Y_pred_svm, Y_test))
    #print(classification_report(Y_test,Y_pred_svm))
```

```python
plt.plot(acc,label='Change of C')
plt.legend()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
acc_decision_tree=[]
c=[]
for i in range(10):
    acc_decision_tree.append([])
for i in range(10):
    dlf=KFold(n_splits=5,random_state=1,shuffle=True)
    dlf.get_n_splits(X)
    for train_index, test_index in dlf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        dlf = DecisionTreeClassifier()
        dlf.fit(X_train,Y_train)
        Y_pred_dt = dlf.predict(X_test)
        acc_decision_tree[i].append(round(dlf.score(X_test, Y_test) * 100, 2))
        c.append(round(dlf.score(X_test, Y_test) * 100, 2))
s3=np.mean(acc_decision_tree)
s3
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import warnings
warnings.filterwarnings("ignore")
acc_logistic=[]
a=[]
for i in range(10):
    acc_logistic.append([])
for i in range(10):
    lr=KFold(n_splits=5,random_state=1,shuffle=True)
    lr.get_n_splits(X)
    for train_index, test_index in lr.split(X):
    #    print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        lr = LogisticRegression()
        lr.fit(X_train,Y_train)
        Y_pred_dt = lr.predict(X_test)
        acc_logistic[i].append(round(lr.score(X_test, Y_test) * 100, 2))
        a.append(round(lr.score(X_test, Y_test) * 100, 2))
        #print(classification_report(Y_test,Y_pred_dt))
s1=np.mean(acc_logistic)
s1
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.ensemble import AdaBoostClassifier
acc_logistic=[]
b=[]
for i in range(10):
    acc_logistic.append([])
for i in range(10):
    lr=KFold(n_splits=5,random_state=1,shuffle=True)
    lr.get_n_splits(X)
    for train_index, test_index in lr.split(X):
    #    print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
        AdaBoost = AdaBoostClassifier(n_estimators=1,learning_rate=100,algorithm='SAMME')
        AdaBoost.fit(X_train,Y_train)
        b.append( AdaBoost.score(X_test,Y_test)*100)
```
